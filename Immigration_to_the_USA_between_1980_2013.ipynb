#### Immigration trend from 1980 to 2013 by Region

```python
def area_plot1(dt,region):
  years=list(map(str,range(1980,2014)))
  data_by_region=dt.groupby(region)[years].sum() 
  data_by_region=data_by_region.transpose() 
  data_by_region.plot(kind='area',figsize=(15,10), alpha=0.5,stacked=False)
  plt.xlabel('Years')
  plt.ylabel('Number of immigrants')
  plt.title('Immigration rate from different regions to the USA from 1980-2013')
```

**Immigration rate from different countries per region to the USA from 1980-2013**

```python
def hist_plot(data,column,region_name,colors=None):
  years=list(map(str,range(1980,2014)))
  data=data[data[column]==region_name]
  data=data[years].transpose()
  count,bin_edges=np.histogram(data,15)
  data.plot(kind='hist',figsize=(15,10), alpha=0.3,stacked=False,color=colors)
  plt.xlabel('Number of immigrants')
  plt.ylabel('Number of Years')
  plt.title('Immigration rate from '+ region_name+ ' to the USA from 1980-2013')
```

```python
def line_plot(data,column=None,region_name=None,colors=None):
  years=list(map(str,range(1980,2014)))
  data=data[data[column]==region_name]
  data=data[years].transpose()
  count,bin_edges=np.histogram(data,15)
  data.plot(kind='line',figsize=(15,10),color=colors)
  plt.xlabel('Years')
  plt.ylabel('Number of immigrants')
  plt.title('Immigration rate from '+ region_name+ ' to the USA from 1980-2013')
```

```python
def area_plot(data,column,region_name,colors=None):
  years=list(map(str,range(1980,2014)))
  data=data[data[column]==region_name]
  data=data[years].transpose()
  count,bin_edges=np.histogram(data,15)
  data.plot(kind='area',figsize=(15,10), stacked=False, alpha=0.3, color=colors)
  plt.xlabel('Years')
  plt.ylabel('Number of immigrants')
  plt.title('Immigration rate from '+ region_name+ ' to the USA from 1980-2013')
```
